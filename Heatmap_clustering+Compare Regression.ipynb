```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
data_In = pd.read_csv("/content/Inline_generated.csv")

# Show dataframe


plt.figure(figsize=(8,8))
correlations_In = data_In.corr()
#correlations = data.corr(method='kendall')
abous_In= abs(correlations_In)

res_In = sns.heatmap(round(abous_In,2), cmap=plt.cm.CMRmap_r, annot=True,
            annot_kws={"size": 15}, vmin=0, vmax=1,)
res_In.set_xticklabels(res_In.get_xmajorticklabels(), fontsize = 12,rotation=20)
res_In.set_yticklabels(res_In.get_ymajorticklabels(), fontsize = 12)

cax_In = res_In.figure.axes[-1]
cax_In.tick_params(labelsize=15)


plt.savefig("Inline_heatmap.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
clustermap_In = sns.clustermap(abous_In, method="complete", cmap=plt.cm.CMRmap_r, annot=True,
               annot_kws={"size": 12}, vmin=0, vmax=1, figsize=(8,8))

#g = sns.clustermap(data=corr_s, annot=True, fmt='d',cmap = "Blues")
clustermap_In.ax_heatmap.set_xticklabels(clustermap_In.ax_heatmap.get_xmajorticklabels(),fontsize = 12)
clustermap_In.ax_heatmap.set_yticklabels(clustermap_In.ax_heatmap.get_ymajorticklabels(),fontsize = 11)

cax_In = clustermap_In.figure.axes[-1]
cax_In.tick_params(labelsize=15)
plt.savefig("Inline_Clustermap.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
